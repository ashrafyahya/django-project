# Django  

Django ist ein Python-Framework, das es erleichtert, Websites mit Python zu erstellen.

Mit Django können Sie Webanwendungen innerhalb weniger Stunden vom Konzept bis zum Start bringen. Django übernimmt einen Großteil des Umstands der Webentwicklung, sodass Sie sich auf das Schreiben Ihrer App konzentrieren können, ohne das Rad neu erfinden zu müssen. 

Django legt großen Wert auf die Wiederverwendbarkeit von Komponenten, auch als DRY bezeichnet (Don't Repeat Yourself), und bietet einsatzbereite Funktionen wie das Login-System, Datenbankverbindung und CRUD-Operationen (Erstellen Lesen, Aktualisieren, Löschen).

Django ist kostenlos & Open Source und wurde 2003 von Lawrence Journal-World erfunden, um dem Kurzfilm gerecht zu werden, Deadlines in der Zeitung zu erfüllen und gleichzeitig die Anforderungen von Erfahrenen Webentwicklern. Die erste öffentliche Veröffentlichung erfolgte im Juli 2005.

## Wie funktioniert Django?

![](./bilder/Django-Verlauf.png)

Django folgt dem **MVT-Designmuster** (Model View Template):

- **Modell**: Die Daten, die Sie präsentieren möchten, meist Daten aus einer Datenbank.
- **View**: Ein Anforderungshandler, der die relevante Vorlage und Inhalte basierend auf der Anfrage des Benutzers zurückgibt.
- **Vorlage**: Eine Textdatei (wie eine HTML-Datei), die das Layout der Webseite enthält, mit einer Logik zur Darstellung der Daten.

## Modell

Das **[Modell](https://www.w3schools.com/django/django_models.php)** liefert Daten aus der Datenbank.  
In Django werden die Daten als Objekt-Relational-Mapping (ORM) geliefert. Das ist eine Technik, die darauf ausgelegt ist, die Arbeit mit Datenbanken zu erleichtern.

Die gebräuchlichste Methode, Daten aus einer Datenbank zu extrahieren, ist SQL. Ein Problem mit SQL bedeutet, dass man ein ziemlich gutes Verständnis der Datenbankstruktur haben muss damit arbeiten zu können.

Django erleichtert mit ORM die Kommunikation mit der Datenbank, ohne komplexe SQL-Anweisungen schreiben zu müssen.  
Die Modelle befinden sich üblicherweise in einer Datei namens `models.py`.  

## View
Eine **[View](https://www.w3schools.com/django/django_views.php)** ist eine Funktion oder Methode, die HTTP-Anfragen als Argumente nimmt, importiert das entsprechende Modell bzw. die entsprechenden Modelle, und findet heraus, welche Daten an die Vorlage gesendet werden sollen, und liefert das Endergebnis zurück.  

Die Views befinden sich üblicherweise in einer Datei namens `iews.py`.  


## Template

Eine **[Template]((https://www.w3schools.com/django/django_templates.php))** ist eine Datei, in der Sie beschreiben, wie das Ergebnis dargestellt werden soll.

Templates sind oft .html Dateien, wobei HTML-Code das Layout einer Webseite beschreibt. Aber es kann auch in anderen Dateiformaten sein, um andere Ergebnisse zu präsentieren. Wir konzentrieren uns allerdings auf .html Dateien.

Django verwendet Standard-HTML, um das Layout zu beschreiben, verwendet aber Django-Tags, um Logik hinzuzufügen:  
```<h1>My Homepage</h1>  
<p>My name is {{ firstname }}.</p>
```  
Die Templates einer Anwendung befinden sich in einem Ordner namens ``templates``.  

## URLs
Django bietet außerdem eine Möglichkeit, sich auf den verschiedenen Seiten einer Website zurechtzufinden.  
Wenn ein Nutzer eine [URL](https://www.w3schools.com/django/django_urls.php) anfordert, entscheidet Django, an welche Ansicht er sie weiterleitet.  

Dies geschieht in einer Datei namens ``urls.py``.
