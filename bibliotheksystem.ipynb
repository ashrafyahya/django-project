## 8) Models implementieren (Kernlogik)


### Datei `library/models.py` komplett ersetzen:
Diese Datei beschreibt die Datenstruktur des Projekts.
Hier legen wir fest, welche Informationen gespeichert werden und welche Regeln fuer Buchungen und Bestand gelten.
Anfaenger sollten sich merken: Models sind das Herz der Datenbanklogik.


```python
from django.conf import settings
from django.core.exceptions import ValidationError
from django.db import models, transaction


class Category(models.Model):
    # Dieses Model speichert Buchkategorien.
    # Beispiele: Programming, Fiction, History.
    name = models.CharField(max_length=100, unique=True)
    # Die Beschreibung ist optional und hilft spaeter im Admin.
    description = models.TextField(blank=True)

    class Meta:
        # Kategorien werden alphabetisch angezeigt.
        ordering = ['name']

    def __str__(self):
        # Diese Methode bestimmt, wie ein Objekt als Text erscheint,
        # z. B. im Django-Admin oder in der Shell.
        return self.name


class Book(models.Model):
    # Grunddaten des Buches.
    title = models.CharField(max_length=200)
    author = models.CharField(max_length=120)

    # Jede Kategorie kann viele Buecher haben.
    # PROTECT verhindert das Loeschen einer Kategorie,
    # solange noch Buecher mit ihr verknuepft sind.
    category = models.ForeignKey(Category, on_delete=models.PROTECT, related_name='books')

    # `total_copies` beschreibt den Gesamtbestand.
    total_copies = models.PositiveIntegerField(default=1)
    # `available_copies` beschreibt, wie viele Exemplare aktuell frei sind.
    available_copies = models.PositiveIntegerField(default=1)

    # Zeitstempel der Erstellung.
    created_at = models.DateTimeField(auto_now_add=True)

    class Meta:
        # Buecher werden alphabetisch nach Titel sortiert.
        ordering = ['title']

    def clean(self):
        # Diese Validierung schuetzt vor unlogischen Daten.
        # Verfuegbar darf nie groesser sein als der Gesamtbestand.
        if self.available_copies > self.total_copies:
            raise ValidationError('Available copies cannot be greater than total copies.')

    def __str__(self):
        # Gut lesbare Darstellung des Buches.
        return f'{self.title} - {self.author}'


class Booking(models.Model):
    # Die moeglichen Statuswerte der Buchung.
    # `TextChoices` ist praktisch, weil Django damit direkt
    # Auswahlfelder und lesbare Labels erzeugen kann.
    class Status(models.TextChoices):
        BOOKED = 'booked', 'Booked'
        CANCELLED = 'cancelled', 'Cancelled'
        BORROWED = 'borrowed', 'Borrowed'
        RETURNED = 'returned', 'Returned'

    # Wer hat das Buch gebucht?
    student = models.ForeignKey(settings.AUTH_USER_MODEL, on_delete=models.CASCADE, related_name='bookings')
    # Welches Buch wurde gebucht?
    book = models.ForeignKey(Book, on_delete=models.CASCADE, related_name='bookings')
    # Aktueller Status der Buchung.
    status = models.CharField(max_length=20, choices=Status.choices, default=Status.BOOKED)
    # Automatische Zeitstempel fuer Nachvollziehbarkeit.
    booked_at = models.DateTimeField(auto_now_add=True)
    updated_at = models.DateTimeField(auto_now=True)

    class Meta:
        # Neueste Buchungen zuerst anzeigen.
        ordering = ['-booked_at']

    def __str__(self):
        # Lesbare Darstellung einer Buchung.
        return f'{self.student.username} - {self.book.title} ({self.status})'

    @classmethod
    def create_booking(cls, student, book):
        # Diese Methode erstellt eine neue Buchung.
        # Wir kapseln die Logik hier, damit sie nicht in Views oder Admin
        # doppelt geschrieben werden muss.
        with transaction.atomic():
            # Das Buch wird gesperrt, damit zwei parallele Buchungen
            # nicht gleichzeitig denselben Bestand verbrauchen.
            locked_book = Book.objects.select_for_update().get(pk=book.pk)

            # Wenn nichts verfuegbar ist, stoppen wir mit Fehlermeldung.
            if locked_book.available_copies < 1:
                raise ValidationError('No copies are currently available.')

            # Ein Nutzer soll dasselbe Buch nicht mehrfach aktiv buchen koennen.
            has_active_booking = cls.objects.filter(
                student=student,
                book=locked_book,
                status__in=[cls.Status.BOOKED, cls.Status.BORROWED],
            ).exists()
            if has_active_booking:
                raise ValidationError('You already have an active booking for this book.')

            # Erst Bestand reduzieren, dann die Buchung speichern.
            locked_book.available_copies -= 1
            locked_book.save(update_fields=['available_copies'])
            return cls.objects.create(student=student, book=locked_book, status=cls.Status.BOOKED)

    def cancel_by_student(self):
        # Diese Methode wird verwendet, wenn ein Studierender selbst storniert.
        # Erlaubt ist das nur direkt nach der Buchung.
        if self.status != self.Status.BOOKED:
            raise ValidationError('Only booked items can be cancelled by students.')
        self._cancel_and_restore_inventory()

    def cancel_by_admin(self):
        # Auch im Admin gilt dieselbe Grundregel:
        # Nur offene Buchungen duerfen storniert werden.
        if self.status != self.Status.BOOKED:
            raise ValidationError('Only booked items can be cancelled.')
        self._cancel_and_restore_inventory()

    def mark_as_borrowed(self):
        # Hier wechselt der Status von reserviert zu ausgeliehen.
        # Der Bestand wird dabei NICHT veraendert, weil er bereits
        # bei der Buchung reduziert wurde.
        if self.status != self.Status.BOOKED:
            raise ValidationError('Only booked items can be marked as borrowed.')
        self.status = self.Status.BORROWED
        self.save(update_fields=['status', 'updated_at'])

    def mark_as_returned(self):
        # Rueckgabe ist nur sinnvoll, wenn das Buch vorher ausgeliehen war.
        if self.status != self.Status.BORROWED:
            raise ValidationError('Only borrowed items can be marked as returned.')

        with transaction.atomic():
            # Buch erneut sperren, damit der Bestand sicher aktualisiert wird.
            locked_book = Book.objects.select_for_update().get(pk=self.book.pk)
            locked_book.available_copies += 1

            # Sicherheit: Der Bestand darf nie ueber dem Gesamtbestand liegen.
            if locked_book.available_copies > locked_book.total_copies:
                locked_book.available_copies = locked_book.total_copies
            locked_book.save(update_fields=['available_copies'])

            # Danach den Status der Buchung auf RETURNED setzen.
            self.status = self.Status.RETURNED
            self.save(update_fields=['status', 'updated_at'])

    def _cancel_and_restore_inventory(self):
        # Diese interne Hilfsmethode vermeidet doppelten Code.
        # Sowohl Student als auch Admin verwenden dieselbe Storno-Logik.
        with transaction.atomic():
            locked_book = Book.objects.select_for_update().get(pk=self.book.pk)
            locked_book.available_copies += 1
            if locked_book.available_copies > locked_book.total_copies:
                locked_book.available_copies = locked_book.total_copies
            locked_book.save(update_fields=['available_copies'])

            self.status = self.Status.CANCELLED
            self.save(update_fields=['status', 'updated_at'])
```

### Statusregeln (wichtig)



| Aktion | Erlaubt von Status | Neuer Status | Bestand (`available_copies`) |
|---|---|---|---|
| Student bucht | - | `booked` | `-1` |
| Student storniert | `booked` | `cancelled` | `+1` |
| Admin storniert | `booked` | `cancelled` | `+1` |
| Admin setzt ausgeliehen | `booked` | `borrowed` | unverändert |
| Admin setzt zurückgegeben | `borrowed` | `returned` | `+1` |


## 9) Views implementieren


### Datei `library/views.py` komplett ersetzen:
Diese Datei enthaelt die Logik hinter den Seiten.
Views holen Daten aus den Models, pruefen Eingaben und uebergeben Ergebnisse an Templates.
Kurz gesagt: Models speichern Daten, Views arbeiten mit diesen Daten.



```python
from django.contrib import messages
from django.contrib.auth.decorators import login_required
from django.core.exceptions import ValidationError
from django.shortcuts import get_object_or_404, redirect, render

from .models import Book, Booking


def home_redirect(request):
    # Diese Funktion ist unsere einfache Startlogik.
    # Wer eingeloggt ist, soll direkt den Buchkatalog sehen.
    # Wer nicht eingeloggt ist, wird zuerst zur Login-Seite geschickt.
    if request.user.is_authenticated:
        return redirect('book_list')
    return redirect('login')


@login_required
def book_list(request):
    # Diese View laedt alle Buecher fuer die Katalog-Seite.
    # `select_related('category')` ist eine kleine Optimierung:
    # Django holt Kategorie-Daten direkt mit und braucht spaeter
    # nicht fuer jedes Buch eine zusaetzliche Datenbankabfrage.
    books = Book.objects.select_related('category').all()

    # `render()` verbindet Daten mit einem HTML-Template.
    return render(request, 'library/book_list.html', {'books': books})


@login_required
def my_bookings(request):
    # Diese View zeigt nur die Buchungen des aktuell eingeloggten Users.
    # Dadurch sieht jeder Studierende nur seine eigenen Daten.
    bookings = Booking.objects.select_related('book').filter(student=request.user)
    return render(request, 'library/my_bookings.html', {'bookings': bookings})


@login_required
def create_booking(request, book_id):
    # Buchungen sollen nur ueber ein Formular mit POST erzeugt werden.
    # So verhindern wir ungewollte Aenderungen durch einfaches Oeffnen einer URL.
    if request.method != 'POST':
        return redirect('book_list')

    # Hole das angeklickte Buch oder liefere 404, falls die ID ungueltig ist.
    book = get_object_or_404(Book, pk=book_id)
    try:
        # Die eigentliche Fachlogik bleibt im Model.
        # Das macht den Code sauberer und leichter wiederverwendbar.
        Booking.create_booking(student=request.user, book=book)
        messages.success(request, 'Buchung erfolgreich erstellt.')
    except ValidationError as exc:
        # Wenn die Buchung fachlich nicht erlaubt ist,
        # zeigen wir die Fehlermeldung fuer den Nutzer an.
        messages.error(request, exc.message)

    # Nach der Aktion zur Buchliste zurueckkehren.
    return redirect('book_list')


@login_required
def cancel_booking(request, booking_id):
    # Auch das Stornieren soll nur per POST erfolgen.
    if request.method != 'POST':
        return redirect('my_bookings')

    # Wichtig: Es darf nur eine Buchung des aktuellen Users geladen werden.
    # So kann niemand fremde Buchungen stornieren.
    booking = get_object_or_404(Booking, pk=booking_id, student=request.user)
    try:
        # Die Fachregel fuer das Storno liegt wieder im Model.
        booking.cancel_by_student()
        messages.success(request, 'Buchung wurde storniert.')
    except ValidationError as exc:
        messages.error(request, exc.message)

    return redirect('my_bookings')
```


## 10) Admin konfigurieren


### Datei `library/admin.py` komplett ersetzen:
Diese Datei bestimmt, wie unsere Models im Django-Admin angezeigt und bearbeitet werden.
Der Admin ist fuer dieses Tutorial wichtig, weil wir dort schnell Kategorien, Buecher und Buchungen verwalten koennen.



```python
from django.contrib import admin
from django.core.exceptions import ValidationError

from .models import Book, Booking, Category


@admin.register(Category)
class CategoryAdmin(admin.ModelAdmin):
    # `list_display` bestimmt, welche Spalten in der Admin-Liste sichtbar sind.
    list_display = ('name',)
    # `search_fields` erlaubt die Suche ueber das angegebene Feld.
    search_fields = ('name',)


@admin.register(Book)
class BookAdmin(admin.ModelAdmin):
    # Hier sehen Admins die wichtigsten Buchinformationen auf einen Blick.
    list_display = ('title', 'author', 'category', 'total_copies', 'available_copies')
    # Filter links im Admin, um z. B. nach Kategorie einzugrenzen.
    list_filter = ('category',)
    # Suche nach Titel oder Autor.
    search_fields = ('title', 'author')


@admin.register(Booking)
class BookingAdmin(admin.ModelAdmin):
    # Zeigt die wichtigsten Daten einer Buchung in der Admin-Liste.
    list_display = ('student', 'book', 'status', 'booked_at', 'updated_at')
    list_filter = ('status',)
    search_fields = ('student__username', 'book__title')

    # Diese Aktionen koennen auf mehrere markierte Buchungen gleichzeitig angewendet werden.
    actions = ('action_cancel', 'action_mark_borrowed', 'action_mark_returned')

    @admin.action(description='Set selected bookings to cancelled')
    def action_cancel(self, request, queryset):
        # Zaehler fuer erfolgreich geaenderte Datensaetze.
        updated = 0
        for booking in queryset:
            try:
                # Wir verwenden bewusst die Model-Methode,
                # damit die Fachlogik nicht doppelt im Admin steht.
                booking.cancel_by_admin()
                updated += 1
            except ValidationError:
                # Unzulaessige Statuswechsel werden hier einfach uebersprungen.
                continue
        self.message_user(request, f'{updated} booking(s) set to cancelled.')

    @admin.action(description='Set selected bookings to borrowed')
    def action_mark_borrowed(self, request, queryset):
        updated = 0
        for booking in queryset:
            try:
                booking.mark_as_borrowed()
                updated += 1
            except ValidationError:
                continue
        self.message_user(request, f'{updated} booking(s) set to borrowed.')

    @admin.action(description='Set selected bookings to returned')
    def action_mark_returned(self, request, queryset):
        updated = 0
        for booking in queryset:
            try:
                booking.mark_as_returned()
                updated += 1
            except ValidationError:
                continue
        self.message_user(request, f'{updated} booking(s) set to returned.')
```

### Datei `templates/library/my_bookings.html`:

## 13) Testdatei für Logiktests

### Datei `library/tests.py` komplett ersetzen:
Diese Datei prueft automatisch, ob unsere wichtigste Logik wirklich funktioniert.
Gerade fuer Anfaenger ist das wichtig: Tests bestaetigen, dass der Bestand bei Buchung, Storno und Rueckgabe korrekt veraendert wird.


```python
from django.contrib.auth import get_user_model
from django.core.exceptions import ValidationError
from django.test import TestCase

from .models import Book, Booking, Category


class BookingWorkflowTests(TestCase):
    # Diese Testklasse prueft die wichtigste Geschaeftslogik unseres Projekts.
    # Jeder Test beantwortet eine konkrete Frage, z. B.:
    # "Wird der Bestand wirklich reduziert, wenn jemand bucht?"

    def setUp(self):
        # `setUp()` laeuft vor JEDEM einzelnen Test neu.
        # So startet jeder Test mit frischen, sauberen Ausgangsdaten.
        self.user = get_user_model().objects.create_user(username='student1', password='test12345')
        self.category = Category.objects.create(name='Programming')
        self.book = Book.objects.create(
            title='Django Basics',
            author='Team Tutorial',
            category=self.category,
            total_copies=3,
            available_copies=3,
        )

    def test_create_booking_reduces_available_copies(self):
        # Dieser Test prueft den Normalfall einer erfolgreichen Buchung.
        Booking.create_booking(self.user, self.book)

        # Mit `refresh_from_db()` laden wir den aktuellen Stand erneut aus der DB.
        self.book.refresh_from_db()

        # Erwartung: 3 verfuegbare Exemplare werden zu 2.
        self.assertEqual(self.book.available_copies, 2)

    def test_cancel_booking_increases_available_copies(self):
        # Zuerst eine Buchung erzeugen, damit es ueberhaupt etwas zu stornieren gibt.
        booking = Booking.create_booking(self.user, self.book)

        # Danach wird die Buchung storniert.
        booking.cancel_by_student()

        self.book.refresh_from_db()
        booking.refresh_from_db()

        # Erwartung: Der Bestand ist wieder beim Ausgangswert.
        self.assertEqual(self.book.available_copies, 3)
        # Erwartung: Der Status wurde korrekt gesetzt.
        self.assertEqual(booking.status, Booking.Status.CANCELLED)

    def test_returned_booking_restores_inventory(self):
        # Dieser Test simuliert den vollstaendigen Ablauf:
        # buchen -> ausleihen -> zurueckgeben.
        booking = Booking.create_booking(self.user, self.book)
        booking.mark_as_borrowed()
        booking.mark_as_returned()
        self.book.refresh_from_db()

        # Nach der Rueckgabe soll der Bestand wieder vollstaendig sein.
        self.assertEqual(self.book.available_copies, 3)

    def test_cannot_book_unavailable_book(self):
        # Hier pruefen wir einen Fehlerfall.
        # Wenn kein Exemplar mehr frei ist, muss Django eine ValidationError werfen.
        self.book.available_copies = 0
        self.book.save(update_fields=['available_copies'])

        with self.assertRaises(ValidationError):
            Booking.create_booking(self.user, self.book)
```
